```python
# Calcular média de monthlyincome por tipo de jornada (standardhours)
media_income_FT_PT = df_limpio.groupby("standardhours")["monthlyincome_$"].mean()

media_income_FT_PT
```

```python
# ————— Funcion para substituir valores ausentes de "standardhours" com base no salário —————

# Calcular la media de salario para empleados de tiempo integral
media_ft = df_limpio[df_limpio["standardhours"] == "Full Time"]["monthlyincome_$"].mean()

# Calcular la media de salario para empleados de medio período
media_pt = df_limpio[df_limpio["standardhours"] == "Part Time"]["monthlyincome_$"].mean()

# Definir la funcion que atribui "Full Time" o "Part Time" com base no salário
def rellenar_tipo_jornada(salario):
    if pd.isna(salario):
        return "Unknown"  # Se el salario esta ausente, no es possible elegir
    elif abs(salario - media_ft) < abs(salario - media_pt):
        return "Full Time"
    else:
        return "Part Time"

# Substituir "Unknown", string vazia ou espaços por NaN para padronizar os ausentes
df_limpio["standardhours"] = df_limpio["standardhours"].replace(["Unknown", "", " "], np.nan)

# Preencher os valores ausentes com a função baseada no salário
df_limpio["standardhours"] = df_limpio["standardhours"].fillna(df_limpio["monthlyincome_$"].apply(rellenar_tipo_jornada))

```

```python
#tanto roledepartment como department, tienen muchos valores nulos. La solución está en hacer 2 pasos:
#rellenar valores faltantes en department, y después rellenamos roledepartment con department + jobrole


# --- PASO 1: RELLENAR DATOS FALTANTES EN 'department' ---

print(f"Valores nulos en 'department' ANTES: {df_definitivo['department'].isnull().sum()}")

# Limpiamos la columna 'department' para que sea consistente (minúsculas, sin espacios)
# Nota: 'jobrole' ya fue limpiado en tu notebook.
df_definitivo['department'] = df_definitivo['department'].str.strip().str.lower()

# Creamos un mapa para inferir el departamento a partir del 'jobrole'.
# Este mapa se basa en la lógica que se observa en los datos que ya tienes.
role_to_dept_map = {
    'sales executive': 'sales',
    'sales representative': 'sales',
    'human resources': 'human resources',
    'research director': 'research & development',
    'research scientist': 'research & development',
    'laboratory technician': 'research & development',
    'manufacturing director': 'research & development',
    'healthcare representative': 'research & development'
    # 'manager' es ambiguo y podría pertenecer a varios departamentos, por lo que lo omitimos del mapa.
}

#Usamos el mapa para obtener los departamentos inferidos
inferred_dept = df_definitivo['jobrole'].map(role_to_dept_map)

# Rellenamos los valores nulos en 'department' con los valores que inferimos.
# La función .fillna() solo actúa sobre los valores NaN, sin sobreescribir los que ya existen.
df_definitivo['department'].fillna(inferred_dept, inplace=True)

print(f"Valores nulos en 'department' DESPUÉS: {df_definitivo['department'].isnull().sum()}")

```

```python
# --- PASO 2: CONSTRUIR LA COLUMNA 'roledepartament' ---

print(f"\nValores nulos en 'roledepartament' ANTES: {df_definitivo['roledepartament'].isnull().sum()}")

# Usamos np.where para construir la columna. Es eficiente y maneja la lógica correctamente.
# Condición: si 'department' NO es nulo.
# Si es verdad: concatena 'jobrole' y 'department'.
# Si es falso: deja el valor como NaN.
df_definitivo['roledepartament'] = np.where(
    df_definitivo['department'].notna(),
    df_definitivo['jobrole'] + ' - ' + df_definitivo['department'],
    np.nan
)

print(f"Valores nulos en 'roledepartament' DESPUÉS: {df_definitivo['roledepartament'].isnull().sum()}")

```
